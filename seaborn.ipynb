# 1. Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

Although, scatterplot can be used for both numerical as well as categorical data, it's mostly used to depict the relationship between the numerical data.

There are two ways to draw a scatter plot in Seaborn

### 1. replot : Figure Level Function
### 2. Scatterplot : Axes Level Function

As we will see these, visualisations can be quite helpful because they use simply and easily understood representations of the data even on the complex data by mapping up to 3 additional variable using the semantics of heu, size and style.

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("fivethirtyeight")
```

```python
#Load an example dataset
#tips = sns.load_dataset('tips')
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

# Plot a Scatter Plot between total_bill and tip

```python
# Create a visualization
sns.relplot( data=data, x="total_bill", y="tip") #kind = 'scatter'... default value is scatter
```

# The hue parameter

```python
sns.relplot(
    data=data,
    x="total_bill", y="tip", 
    hue="smoker")
```

# The style parameter

```python
sns.relplot(
    data=data,
    x="total_bill", y="tip", 
    hue="smoker", style = "sex")
```

# Size Parameter

```python
sns.relplot(
    data=data,
    x="total_bill", y="tip", 
    hue="smoker", style = "sex", size = "size")
```

```python
sns.relplot(
    data=data,
    x="total_bill", y="tip", 
    hue="smoker", style = "sex", size = "size", sizes = (15,200)) #sizes normalizes the area occupied by this range
```

# using the scatterplot function

```python
#repalce replot by scatterplot
sns.scatterplot(
    data=data,
    x="total_bill", y="tip", 
    hue="smoker")
```

# Violin Plots

Violin plots are a method of ploting numeric data and can be considered a combination of the box plot with a kernel density plot or also called PDF (Probability density function)

The default violinplot represents a distribution two ways: a patch showing a symmetric kernel density estimate (KDE), and the quartiles / whiskers of a box plot:

Violin plots are often used to compare the distribution of a given variable across some categories

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', data=tips, split=True)
```
